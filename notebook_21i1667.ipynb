```python
def clean_description(description):
    if isinstance(description, str):
        tokens = word_tokenize(description)
        stop_words = set(stopwords.words('english'))
        filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
        stemmer = PorterStemmer()
        lemmatizer = WordNetLemmatizer()
        stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
        lemmatized_tokens = [lemmatizer.lemmatize(word) for word in stemmed_tokens]
        cleaned_description = ' '.join(lemmatized_tokens)
        return cleaned_description
    else:
        return None

def clean_descriptions(descriptions):
    cleaned_descriptions = [clean_description(description) for description in descriptions]
    return cleaned_descriptions

```

```python
data['Description'] = clean_description(data['description'])
```

```python
def clean_reviews(reviews):
    cleaned_reviews = []
    for review_list in reviews:
        cleaned_review_list = [clean_description(review) for review in review_list]
        cleaned_reviews.append(cleaned_review_list)
    return cleaned_reviews

data['Reviews'] = data['reviews'].apply(clean_description)
```

```python
def clean_review(review):
    if isinstance(review, str):
        tokens = word_tokenize(review)
        stop_words = set(stopwords.words('english'))
        filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
        stemmer = PorterStemmer()
        lemmatizer = WordNetLemmatizer()
        stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
        lemmatized_tokens = [lemmatizer.lemmatize(word) for word in stemmed_tokens]
        cleaned_review = ' '.join(lemmatized_tokens)
        return cleaned_review
    else:
        return None

# Apply clean_review function to each review in the 'reviews' column
data['cleaned_reviews'] = data['reviews'].apply(lambda reviews_list: [clean_review(review) for review in reviews_list])

```

```python
def clean_text(text):
    if isinstance(text, str):
        return text.lower()
    elif text == 'Male':
        return 'Men'
    else:
        return text


columns_to_clean = ['title', 'gender', 'category', 'sub_category', 'brand_name']
for column in columns_to_clean:
    data[column] = data[column].apply(clean_text)
```

```python
import mysql.connector

# Establish connection to MySQL database
mydb = mysql.connector.connect(
    host='localhost',
    port=3306,
    user='root',
    password='root123',
    database='amazonn'
)

cursor = mydb.cursor()

# Function to insert categories into the database
def insert_categories(conn, df):
    for category in df['category'].unique():
        # Check if the category already exists
        cursor.execute("SELECT COUNT(*) FROM categories WHERE category_name = %s", (category,))
        count = cursor.fetchone()[0]
        if count == 0:
            # Insert the category if it doesn't exist
            cursor.execute("INSERT INTO categories (category_name) VALUES (%s)", (category,))
        else:
            print(f"Skipping duplicate category: {category}")
    conn.commit()

# Function to retrieve category ID from the database
def get_category_id(conn, category):
    cursor.execute("SELECT id FROM categories WHERE category_name = %s", (category,))
    category_id = cursor.fetchone()[0]
    return category_id

# Function to insert products into the database
def insert_products(conn, df):
    for index, row in df.iterrows():
        # Check if product already exists
        cursor.execute("SELECT COUNT(*) FROM products WHERE asin = %s", (row['asin'],))
        count = cursor.fetchone()[0]
        if count == 0:
            # Retrieve category ID
            category_id = get_category_id(conn, row['category'])
            # Define the insert query
            insert_query = """
            INSERT INTO products (URL, title, asin, ratings, category_id, sub_category, brand_name, price, product_description)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
            """
            # Handle missing values
            row_values = []
            for col in ['URL', 'title', 'asin', 'ratings', 'sub_category', 'brand_name', 'price', 'Description']:
                value = row[col]
                if pd.isnull(value):
                    value = None  # Replace NaN with None (NULL in SQL)
                row_values.append(value)
            # Extract values from DataFrame row
            values = (
                row_values[0], row_values[1], row_values[2], row_values[3],
                category_id, row_values[4], row_values[5], row_values[6], row_values[7]
            )
            # Execute the insert query
            cursor.execute(insert_query, values)
        else:
            print(f"Skipping duplicate product ASIN: {row['asin']}")
    conn.commit()

# Function to insert reviews into the database
def insert_reviews(conn, df, cursor):
    for index, row in df.iterrows():
        # Retrieve product ID
        cursor.execute("SELECT product_id FROM products WHERE asin = %s", (row['asin'],))
        product_row = cursor.fetchone()
        if product_row:
            product_id = product_row[0]
            # Iterate over each review in the tuple
            for review in row['cleaned_reviews']:
                # Insert review
                cursor.execute("INSERT INTO reviews (product_id, review) VALUES (%s, %s)", (product_id, review))
        else:
            print(f"No product found for ASIN: {row['asin']}")

    conn.commit()


# Insert categories
insert_categories(mydb, data)

# Insert products
insert_products(mydb, data)

# Insert reviews
insert_reviews(mydb, data, cursor)

# Close cursor and connection
cursor.close()
mydb.close()
```
